```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['target'] = iris.target
data.describe()
```

```python
# 基于透视表完成简单数据统计分析
data.pivot_table(data,index='target')
```

```python
# 可视化
data.plot.scatter(x='petal length (cm)',y='petal width (cm)',
          c='target',cmap='jet')
plt.show()
```

```python
data.plot.scatter(x='sepal length (cm)',y='sepal width (cm)',
          c='target',cmap='jet')
plt.show()
```

**逻辑回归**
+ 逻辑回归分类模型是一种基于回归思想实现分类业务的分类模型
+ 逻辑回归做二元分类的核心思想为：
1. 针对输出为{0,1}的已知训练样本训练一个回归模型，使得训练样本的预测输出限制在(0,1)的数值区间
2. 使得原类别为0的样本输出更接近于0，使得原类别为1的样本输出更接近于1

```python
# 逻辑回归方程
xs = np.linspace(-10,10,200)
ys = 1 / (1 + np.exp(-xs))
plt.grid(linestyle='--')
plt.plot(xs,ys)
plt.show()
```
